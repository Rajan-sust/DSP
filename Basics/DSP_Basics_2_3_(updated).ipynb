```
import numpy as np
import matplotlib.pyplot as plt
```

```
t = np.arange(0.0,2.0,0.05)
```

$$x(t) = A\sin(2\pi f t + phase)$$

```
def create_sinusoidal(A, f, phase = 0):
  y = A * np.sin(2.0 * np.pi* f * t + phase)
  return y
```

**Create a Sinusoidal Signal $x(t)$ and 3 different plotting in subplots**

```
y = create_sinusoidal(1,1)

fig = plt.figure()

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)

ax1.plot(t,y)
ax2.stem(t,y,markerfmt='.')
ax3.scatter(t,y,marker='.')

fig.show()
```

**Create 3 Sinusoidal Signal $x1(t),x2(t),x3(t)$ then combine and plot the signal**

```
y1 = create_sinusoidal(1,1)
y2 = create_sinusoidal(2,5)
y3 = create_sinusoidal(7,9)
```

```
plt.subplot(2,2,1)
plt.plot(t,y1,c='g')

plt.subplot(2,2,2)
plt.plot(t,y2,c='b')

plt.subplot(2,2,3)
plt.plot(t,y3,c='k')

plt.show()
```

```
plt.plot(t,y1+y2+y3)
plt.title('Composite Signal')
plt.show()
```
