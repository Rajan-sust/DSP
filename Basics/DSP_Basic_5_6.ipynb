```
import numpy as np
import matplotlib.pyplot as plt
```

```
t = np.arange(0.0,2.0,0.00001)
```

$$x(t) = A\sin(2\pi f t + phase)$$

```
def create_sinusoidal(t, A, f, phase = 0):
  y = A * np.sin(2.0 * np.pi* f * t + phase)
  return y
```

```
def create_composite(t, A1, f1, A2, f2, A3, f3):
  y = create_sinusoidal(t, A1, f1) + create_sinusoidal(t, A2, f2) + create_sinusoidal(t, A3, f3)
  return y
  
```

```
composite_signal = create_composite(t, 1, 5, 1, 10, 1, 15)

```

```
plt.plot(t,composite_signal)
plt.title('Composite Signal')
plt.show()
```
