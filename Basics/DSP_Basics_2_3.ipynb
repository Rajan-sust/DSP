```
import matplotlib.pyplot as ploter
import numpy as np
```

$$x(t) = A\sin(2\pi f t + phase)$$

```
def create_sinusoidal(A, f, phase = 0):
  t = np.arange(0.0,2.0,0.01)
  y = A * np.sin(2.0 * np.pi* f * t + phase)
  return y
  
```

```
def plot_signal(y, title):
  t = np.arange(0.0,2.0,0.01)
  ploter.plot(t,y)
  ploter.title(title)
  ploter.xlabel('t in x-axix')
  ploter.ylabel('x(t) in y-axis')
  ploter.show()
  
```

**Create a Sinusoidal Signal $x(t)$ and plot the signal**

```
y = create_sinusoidal(1,1)
plot_signal(y,'A Sine Signal')
```

**Create 3 Sinusoidal Signal $x1(t),x2(t),x3(t)$ then combine and plot the signal**

```
y = create_sinusoidal(1,1) + create_sinusoidal(1.1,1) + create_sinusoidal(1,3)
plot_signal(y,'A Combined Signal')
```
