```
import librosa
import librosa.display
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
from scipy.io import wavfile
from scipy import signal
```

### MFCC (Mel Frequency Cepstrum Coefficients)

```
!wget https://storage.googleapis.com/deepmind-media/pixie/making-music/sample_1.wav
```

```
y, sr = librosa.load('sample_1.wav')
```

```
mfccs = librosa.feature.mfcc(y=y, sr=sr)
```

```
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show()
```

### Filtering a wav file

```
"""
Human voice frequency: 165 to 255
"""
def filter_wav():
  sample_rate, waveform = wavfile.read("sample_1.wav")
  sos = signal.butter(10, [190,200], 'bandpass', fs=sample_rate, output='sos')
  filtered = signal.sosfilt(sos, waveform)
  fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
  ax1.plot(waveform)
  ax1.set_title('Original', fontsize=15)
  ax2.plot(filtered)
  ax2.set_title('After 190-200 Hz bandpass filter', fontsize=15)
  ax2.set_xlabel('n (Samples)',fontsize=15)
  plt.tight_layout()
  plt.show()
```

```
filter_wav()
```

### Image Reshape, Flat

```
def img_reshape():
  path = 'gray.jpeg'
  image = mpimg.imread('gray.jpeg')
  h,w = image.shape
  reshaped_image = image.reshape((h*2,w//2))
  fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(7, 5))
  ax1.set_title('Original')
  ax1.imshow(image, cmap='gray')
  ax2.set_title('Reshaped')
  ax2.imshow(reshaped_image, cmap='gray')
```

```
img_reshape()
```

```
def rgb_to_gray():
  path = 'flower_photos/dandelion/16375088191_2bf2916b53.jpg'
  image = mpimg.imread('flower_photos/dandelion/16375088191_2bf2916b53.jpg')
  gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
  fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(7, 7))
  ax1.set_title('Original Shape: {}'.format(image.shape))
  ax1.imshow(image, cmap='gray')
  ax2.set_title('Gray Shape: {}'.format(gray_image.shape))
  ax2.imshow(gray_image, cmap='gray')

```

```
rgb_to_gray()
```
