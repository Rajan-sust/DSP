```
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

```
def create_sinusoidal(A, f, t, phase = 0):
  x = A * np.sin(2.0 * np.pi * f * t + phase)
  return x
```

### Low Pass Filter 

```
def low_pass_filter():
  t = np.linspace(0, 1, 1000, False)
  sig = create_sinusoidal(1,7,t,0) + create_sinusoidal(1,20,t,0)
  """
  Order = 10
  Critical Frequency = 11
  Filter Options: {‘lowpass’, ‘highpass’, ‘bandpass’, ‘bandstop’}
  Digital Filter Sampling rate = 1000
  """
  sos = signal.butter(10, 11, 'lowpass', fs=1000, output='sos')
  filtered = signal.sosfilt(sos, sig)
  fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
  ax1.plot(t, sig)
  ax1.set_title('7 Hz and 20 Hz sinusoids', fontsize=15)
  ax2.plot(t, filtered)
  ax2.set_title('After 7 Hz low-pass filter', fontsize=15)
  ax2.set_xlabel('Time [seconds]',fontsize=15)
  plt.tight_layout()
  plt.show()
```

```
low_pass_filter()
```

### High Pass Filter

```
def high_pass_filter():
  t = np.linspace(0, 1, 1000, False)
  sig = create_sinusoidal(1,10,t,0) + create_sinusoidal(1,20,t,0)
  """
  Order = 10
  Critical Frequency = 15
  Filter Options: {‘lowpass’, ‘highpass’, ‘bandpass’, ‘bandstop’}
  Digital Filter Sampling rate = 1000
  """
  sos = signal.butter(10, 15, 'highpass', fs=1000, output='sos')
  filtered = signal.sosfilt(sos, sig)
  fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
  ax1.plot(t, sig)
  ax1.set_title('10 Hz and 20 Hz sinusoids', fontsize=15)
  ax2.plot(t, filtered)
  ax2.set_title('After 15 Hz high-pass filter', fontsize=15)
  ax2.set_xlabel('Time [seconds]',fontsize=15)
  plt.tight_layout()
  plt.show()
```

```
high_pass_filter()
```

### Band Pass Filter

```
def band_pass_filter():
  t = np.linspace(0, 1, 1000, False)
  sig = create_sinusoidal(1,25,t,0) + create_sinusoidal(1,35,t,0) + create_sinusoidal(1,45,t,0)
  """
  Order = 10
  Critical Frequency = [33,37]
  Filter Options: {‘lowpass’, ‘highpass’, ‘bandpass’, ‘bandstop’}
  Digital Filter Sampling rate = 1000
  """
  sos = signal.butter(10, [33,37], 'bandpass', fs=1000, output='sos')
  filtered = signal.sosfilt(sos, sig)
  fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
  ax1.plot(t, sig)
  ax1.set_title('25 Hz, 35 Hz and 45 Hz sinusoids', fontsize=15)
  ax2.plot(t, filtered)
  ax2.set_title('After [33,37] Hz band-pass filter', fontsize=15)
  ax2.set_xlabel('Time [seconds]',fontsize=15)
  plt.tight_layout()
  plt.show()
```

```
band_pass_filter()
```
